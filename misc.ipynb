```python
import numpy as np
import matplotlib.pyplot as plt

A = 2
k = 1
x = np.linspace(-A, A, 400)

V_x = 0.5 * k * x**2
rho_x = 1 / (np.pi * np.sqrt(A**2 - x**2))

fig, ax1 = plt.subplots()
ax1.plot(x, V_x, 'b-', label=r"$V(x) = \frac{1}{2}kx^2$")
ax1.set_xlabel("x")
ax1.set_ylabel("V(x)", color="b")
ax1.tick_params(axis="y", labelcolor="b")

ax2 = ax1.twinx()
ax2.plot(x, rho_x, 'r--', label=r"$\rho(x) = \frac{1}{\pi \sqrt{A^2 - x^2}}$")
ax2.set_ylabel("(x)", color="r")
ax2.tick_params(axis="y", labelcolor="r")

ax1.legend(loc="upper left")
ax2.legend(loc="upper right")
plt.title("Plots of $V(x)$ and $\\rho(x)$")
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

def V_eff(r):
    return (1 / (2 * r**2)) - (2 / (3 * r**(3/2)))

V_values = V_eff(r)

plt.figure(figsize=(8, 5))
plt.plot(r, V_values, label=r'$V_{\text{eff}}(r)$', color='r')
plt.xlabel(r'$r$')
plt.ylabel(r'$V_{\text{eff}}$')
plt.title('Effective Potential vs r')
plt.xlim(0, 5)
plt.ylim(-2, 2)
plt.legend()
plt.grid()
plt.show()

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Define the second-order ODE as a system of first-order ODEs
def harmonic_oscillator(t, y, omega):
    return [y[1], -omega**2 * y[0]]

# Parameters
omega = 1.0  # Angular frequency
y0 = [1.0, 0.0]  # Initial conditions: [y(0), y'(0)]
t_span = (0, 10)  # Time span for the solution
t_eval = np.linspace(t_span[0], t_span[1], 300)  # Time points where the solution is evaluated

# Solve the ODE
sol = solve_ivp(harmonic_oscillator, t_span, y0, args=(omega,), t_eval=t_eval)
plt.figure(figsize=(8, 5))
plt.plot(sol.t, sol.y[0], label=r'$y(t)$', color='b')
plt.xlabel(r'$t$')
plt.ylabel(r'$y$')
plt.title('Solution of the Harmonic Oscillator')
plt.legend()
plt.grid()
plt.show()
```
