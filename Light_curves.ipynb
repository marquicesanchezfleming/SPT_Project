```python
from astropy.table import Table
import matplotlib.pyplot as plt
from pandas import DataFrame
```

## 90 GHz Light Curve

```python
dat = Table.read('/Users/Djslime07/Downloads/lightcurve_90GHz_filtered.fits', format='fits')
obj_id = dat['id'][0]
dates90 = dat['dates_ave'][0]
flux90 = dat['flux_SCI'][0]
fluxwgt90 =  dat['flux_WGT'][0]
df = dat.to_pandas()

plt.figure(figsize=(10, 5))
plt.errorbar(df['dates_ave'][0], df['flux_SCI'][0], yerr=df['flux_WGT'][0], fmt='o', label='Flux (90 GHz)')
plt.xlabel('Date')
plt.ylabel('Flux (mK)')
plt.title(f"Light Curve of {obj_id} at 90 GHz")
plt.legend()
plt.show()
```
